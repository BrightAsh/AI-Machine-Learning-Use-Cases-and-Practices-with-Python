```python
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil
import requests

# 1. 현재 노트북 이름 가져오기
colab_url = "http://172.28.0.12:9000/api/sessions"
sessions = requests.get(colab_url).json()
notebook_name = sessions[0]["name"]   # 첫 번째 세션 기준

print("현재 노트북 이름:", notebook_name)

# 2. 경로 설정
src = f"/content/drive/MyDrive/Colab Notebooks/{notebook_name}"
dst_dir = "/content/drive/MyDrive/AI_Machine_Learning_Use_Cases_and_Practices_with_Python/AI-Machine-Learning-Use-Cases-and-Practices-with-Python"
dst = os.path.join(dst_dir, notebook_name)

# 3. 파일 이동
if os.path.exists(src):
    shutil.move(src, dst)
    print(f"✅ {notebook_name} 이동 완료 → {dst}")
else:
    print(f"⚠️ 파일을 찾을 수 없습니다: {src}")
```
